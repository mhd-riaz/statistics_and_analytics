Find corelation and covariance coefficients values for the given features

Hrs: [1.5, 2.2, 4, 5, 7] <br>
GPA: [3, 3.8, 5.6, 8.6, 9.2]

```python
import numpy as np # for numeric computations
import pandas as pd # for dataframes
import matplotlib.pyplot as plt # for plots/graphs
from scipy import stats # for stastical operations
import random # for random
```

```python
# step 1: conver to np arrays
hrs = np.array([1.5, 2.2, 4, 5, 7])
gpa = np.array([3, 3.8, 5.6, 8.6, 9.2])
```

```python
# step 2: calculate correlation
corr = np.corrcoef(gpa, hrs)[0,1]
print("correlation coeff: ", corr)
```

```python
# step 3: calculate covariance
cov = np.cov(gpa, hrs)[0,1]
print("covariance coeff: ", cov)
```

---

Using actual formulas

```python
# mean
n = len(gpa)
gpa_mean = sum(gpa)/n
hrs_mean = sum(hrs)/n

print("mean gpa: ", gpa_mean)
print("mean hrs: ", hrs_mean)
```

```python
# calulate mean difference
gpa_diff = [float(y) - gpa_mean for y in gpa]
hrs_diff = [float(x) - hrs_mean for x in hrs]
```

```python
# product of diff
prod_diff = []
for i in range(0, len(gpa_diff)):
    prod_diff.append(gpa_diff[i]*hrs_diff[i])
print(prod_diff)
```

```python
# finding co-variance

cov= sum(prod_diff)/(n-1)
print(cov)
```

```python
# finding co-relation
import math

denominator = math.pow(sum(gpa_diff), 2) * math.pow(sum(hrs_diff), 2)
corr = cov/math.sqrt(denominator)
print("correlation: ", corr)
```
