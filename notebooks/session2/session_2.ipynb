### Question 2

If the letters of the word "AABRAAKAADAABRAA" are arranged at random, find the probability that, all 10 A occur together?

````python
word = "AABRAAKAADAABRAA"

# Find total number of possible
len_of_word = len(word)

# Find per char count
no_a_char = word.count("A")
no_b_char = word.count("B")
no_d_char = word.count("D")
no_r_char = word.count("R")
no_k_char = word.count("K")

# factorials
a_fact = factorial(no_a_char)
b_fact = factorial(no_b_char)
d_fact = factorial(no_d_char)
k_fact = factorial(no_k_char)
r_fact = factorial(no_r_char)


# Find different characters
diffCharCount = len(["A", "B", "D", "R", "K"])

# total possibilities
total = factorial(len_of_word)/(a_fact*b_fact*d_fact*k_fact*r_fact)

# now find combinations of 10 A's together i.e. 'A' favourable
a_favourable = factorial(no_a_char + no_b_char + no_d_char + no_k_char + no_r_char) / (b_fact * d_fact * k_fact * r_fact)

print(total, a_favourable)


# now find probability of a favourable outcomes
a_probability = a_favourable/total

print("Probability of 10 A's together are: ", a_probability)
````

````python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from math import factorial
import random

# Define the target word
target_word = "AABRAAKAADAABRAA"

# Step 1: Analyze the frequency of each letter
# We need to know how many times each character repeats to use the multiset permutation formula.
char_counts = {}
for char in target_word:
    char_counts[char] = char_counts.get(char, 0) + 1

# Extracting the counts for the variables
# Context: We have 10 'A's, 2 'B's, 2 'R's, 1 'K', and 1 'D'
n_total = len(target_word)  # Total number of letters (16)
count_A = char_counts['A']
count_B = char_counts['B']
count_R = char_counts['R']
count_K = char_counts['K']
count_D = char_counts['D']

# Step 2: Calculate the Total Number of Arrangements (Sample Space)
# Permutations: This is a technical jargon for the "total number of ways you can arrange items." 
# Imagine you have a set of colored balls; permutations tell you how many unique sequences you can form.
# Formula: n! / (n1! * n2! * ... * nk!)
# Where n! (factorial) is the product of all positive integers up to n.
total_denominator = (factorial(count_A) * factorial(count_B) * factorial(count_R) * factorial(count_K) * factorial(count_D))
total_possible_outcomes = factorial(n_total) // total_denominator

# Step 3: Calculate the Favorable Arrangements
# Logic: We want all 10 'A's to occur together. 
# We treat the block of ten "AAAAAAAAAA" as a single "super-item".
# Now we are arranging: (1 Block of A) + (2 B's) + (2 R's) + (1 K) + (1 D)
# Total items to arrange now = 1 + 2 + 2 + 1 + 1 = 7
n_favorable_items = 1 + count_B + count_R + count_K + count_D

# Formula for arranging these 7 items: 7! / (1! * 2! * 2! * 1! * 1!)
favorable_denominator = (factorial(1) * factorial(count_B) * factorial(count_R) * factorial(count_K) * factorial(count_D))
favorable_outcomes = factorial(n_favorable_items) // favorable_denominator

# Step 4: Calculate the Probability
# Probability = (Number of Successful Outcomes) / (Total Number of Possible Outcomes)
probability = favorable_outcomes / total_possible_outcomes

# Output the results
print(f"Total letters in word: {n_total}")
print(f"Total unique arrangements: {total_possible_outcomes}")
print(f"Arrangements with all 'A's together: {favorable_outcomes}")
print(f"The probability is: {probability:.10f} or 1/{total_possible_outcomes // favorable_outcomes}")
````
