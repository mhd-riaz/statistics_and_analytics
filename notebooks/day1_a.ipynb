# 1. Measures of central tendancy for the given data points

Mean, Median, Mode

# 2. Measures of spread

Range, Variance, SD, Quantiles, IQR, 5 member summary, construct box plot(to locate outliers)

DATASET:

[10, 3, 1, 6, 2, 3, 4, 2, 3, 4]

```python
dataset = [10, 3, 1, 6, 2, 3, 4, 2, 3, 4]
n = len(dataset)
```

```python
# using libraries
import numpy as np
from scipy import stats
from matplotlib import pyplot as ptl
```

```python
# Mean
mean = sum(dataset)/n
print("Mean: ",mean)
```

```python
# Using library
l_mean = np.mean(dataset)
print("mean using lib:", l_mean )
```

```python
# Median
median = (n + 1)/2
print("Median: ", median)
```

```python
# Using library
l_median = np.median(dataset)
print("median using lib:", l_median )
```

```python
# Mode
count = {}
maxValue = 0
mode = 0
for i in dataset:
    count[i] = dataset.count(i)
# print(count)

for k, v in count.items():
    if v > maxValue:
        maxValue=v
        mode=k
print("Mode: ", mode)
```

```python
# Using library
l_mode = stats.mode(dataset)[0]
print("mode using lib:", l_mode )
```

```python
# Range

mininum = min(dataset)
maximum = max(dataset)

print("Range: ", mininum, " - ",maximum)
```

```python
# Variance
import math


sq_diff = []
for i in dataset:
    sq_diff.append((i - mean) ** 2)

sq_diff_sum= sum(sq_diff)
variance = sq_diff_sum * (1/(n-1))
print("variance: ", variance)
```

```python
# Using library
l_variance = stats.variation(dataset)
print("variance using lib:", l_mean )
```

```python
# Standard Deviation
sd = math.sqrt(variance)

print("Standard Deviation: ", sd)
```

```python
# quantiles

q1=0.25 * (n+1)
q2=0.50 * (n+1)
q3=0.75 * (n+1)

print("first quantile Q1: ", q1)
print("second quantile Q2: ", q2)
print("third quantile Q3: ", q3)
```

```python
# IQR

iqr = q3 - q1
lb = q1 - 1.5 * iqr
ub = q3 + 1.5 * iqr
print("lower bound: ", lb)
print("upper bound: ", ub)
```

```python
# 5 member summary

print("5 Member summary")
print("---------------------")
print("1. Min: ", min(dataset))
print("2. Max: ", max(dataset))
print("3. Q1: ", q1)
print("4. Q2: ", q2)
print("5. Q3: ", q3)
```
