```python
import numpy as np # for numeric computations
import pandas as pd # for dataframes
import matplotlib.pyplot as plt # for plots/graphs
from scipy import stats # for stastical operations
import random # for random
```

```python
# to have same randomization
random.seed(42)
```

```python
# dataset for weak relation
# sales = [165, 182, 140, 153, 172, 168, 190, 124, 187, 148]
# hours = [7, 8, 4, 6, 3, 2, 4, 9, 11, 12]

# dataset for strong relation
sales = [165, 182, 140, 193, 172, 168, 174, 124, 148, 148]
hours = [7, 8.5, 8, 10, 9, 8, 8.5, 7.5, 9.5, 8.5]
```

Step 1. convert as numpy array

```python
salesArr = np.array(sales, dtype=int)
hrArr = np.array(hours,dtype=float)
```

Step 2. Create data frame

```python
df = pd.DataFrame({"Sales": salesArr, "Hours":hrArr})
df
```

Step 3: Sampling

```python
sample = [int(x) for x in random.sample(list(sales), 6)]
sample
```

```python
meanSales = np.mean(sample)
medianSales = np.median(sample)
modeSales_ = stats.mode(sample, keepdims=True) # keepdims - preserve's dimension
modeSales = int(modeSales_[0])

print("Mean sales:  ", meanSales)
print("Median sales:  ", medianSales)
print("Mode sales:  ", modeSales)
```

Step 4: Dispersion spread

```python
variance_sales = np.var(salesArr, ddof=1)
std_sales=np.std(salesArr,ddof=1)
iqr_sales=np.percentile(salesArr,75) - np.percentile(salesArr,25)

print("Variance of sales: ", variance_sales)
print("STD of sales: ", std_sales)
print("IQR of sales: ", iqr_sales)
```

Co-relation coeffifient

```python
corr = np.corrcoef(salesArr, hrArr)[0,1] # limit values to 0 and 1
print("Correlation b/w sales and hours: ", corr)
```

---

Calculating Skewness & kurtosis

```python
# how much skewness
skew_sales = stats.skew(salesArr)
print("skew in sales: ",skew_sales)
```

```python
# if tails are light/heavy using kurtosis
kurt_sales = stats.kurtosis(salesArr)
print("kurtosis in sales: ", kurt_sales)
```

Plotting

```python
# 1. scatter plot
# used for seeing correlation b/w x and y

plt.scatter(hrArr, salesArr)
plt.xlabel("Working hours") # should be independant variable
plt.ylabel("Sales") # should be dependant variable
plt.title("Hours vs Sales")
plt.show
```

```python
# 2. Box plot

plt.boxplot(salesArr)
plt.title("Sales distribution")
plt.show
```
